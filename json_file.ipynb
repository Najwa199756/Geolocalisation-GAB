```python
import pandas as pd 
import  numpy as np
import matplotlib.pyplot as plt
import branca
import folium
from folium import features
import seaborn as sns
```

```python
df=pd.read_csv('/Users/ok/Downloads/data_gab_synthetique22.csv')
df.columns
df.dtypes
df.shape
```

```python
#garder que les colonnes suivantes: lat,longi,region,montant
df.drop(['id','année','mois','minutes','day','heure','num_operations'], axis = 1, inplace = True)
df
```

```python
#la somme total retirée de chaque gab
c=df.groupby('id_gab')['montant'].sum()
c.shape
```

```python
#convertir la serie en dataframe 
df1=pd.DataFrame({'id_gab':c.index, 'montant_total':c.values})
df1.columns
df1
```

```python
#une fois les montants totaux calculés , on supprime la colonne montant
df.drop(['montant'], axis = 1, inplace = True)
```

```python
#eliminer les doublons 
df2=df.drop_duplicates(subset='id_gab', keep='last')
df2
```

```python
#merger les deux df (notre df avec les donnees calcules )
coord=df2.merge(df1, on=('id_gab'))
coord       
```

```python
coord.to_csv('csv_file', index=False, encoding='utf-8')
```
