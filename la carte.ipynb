```python
import pandas as pd 
import  numpy as np
import matplotlib.pyplot as plt
import branca
import folium
import json
from folium import features
```

```python
df   =  pd.read_csv('/Users/ok/Downloads/data_gab_synthetique22.csv')
```

```python
geo1 = json.load(open(("/Users/ok/Desktop/file.json")))
```

```python
coords = (30,-7.5)
map = folium.Map(location=coords, tiles='OpenStreetMap', zoom_start=6.05)
```

```python

for i in range(len(geo1)):
    tooltip = f"<b>id_gab:</b>{geo1[i]['id_gab']}</br><b>{geo1[i]['region']}</b>"
    folium.Marker([geo1[i]['latitude'], geo1[i]['longtitude']],
                        raduis=float(geo1[i]['montant_total'])/100,
                        color = 'crimson',
                        fill = True,
                        fill_color = 'crimson',
                        popup=f"<b>id_gab:</b>{geo1[i]['id_gab']}</br><b>{geo1[i]['region']}</b>" ,
                        tooltip=tooltip).add_to(map)
    folium.CircleMarker([geo1[i]['latitude'], geo1[i]['longtitude']],
                        raduis=float(geo1[i]['montant_total'])/100,
                        color = 'crimson',
                        fill = True,
                        fill_color = 'crimson',
                        popup=f"<b>id_gab:</b>{geo1[i]['id_gab']}</br><b>{geo1[i]['region']}</b>" ,
                        tooltip=tooltip).add_to(map)
map
```
